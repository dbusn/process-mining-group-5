# Correlation matrix

```python
f, ax = plt.subplots(figsize=(15, 15))
corr = df_train.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool),
            square=True, ax=ax, annot=True, fmt=".1")
```
