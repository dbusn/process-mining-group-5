## 2017 dataset attributes

1. **Action** - 'Created', 'statechange', 'Deleted', 'Obtained', 'Released'

2. **org:resource** -  149 unique User names

3. **concept:name** - 'A_Create Application', 'A_Submitted', 'W_Handle leads', 'W_Complete application', 'A_Concept', 'A_Accepted', 'O_Create Offer', 'O_Created', 'O_Sent (mail and online)', 'W_Call after offers', 'A_Complete', 'W_Validate application', 'A_Validating', 'O_Returned', 'W_Call incomplete files', 'A_Incomplete', 'O_Accepted', 'A_Pending', 'A_Denied', 'O_Refused', 'O_Cancelled', 'A_Cancelled', 'O_Sent (online only)', 'W_Assess potential fraud', 'W_Personal Loan collection', 'W_Shortened completion '  --> ***much more clear than in 2012 dataset!!***
  
4. **EventOrigin** - 'Application', 'Workflow', 'Offer'

5. **EventID** - format:Application_652823628 - every entry unique - key

6. **lifecycle:transition** - 'complete', 'schedule', 'withdraw', 'start', 'suspend', 'ate_abort', 'resume' --> ***2017 has 7 states, 2012 only 3***
       
7. **time:timestamp**

8. **case:LoanGoal** - what is the loan for (14 unique values)

9. **case:ApplicationType** - 'New credit', 'Limit raise'

10. **case:concept:name** - (probably) identifies whole individual cases

11. **case:RequestedAmount**

12. **FirstWithdrawalAmount** - only where request accepted - else, NaN

13. **NumberOfTerms** - only where request accepted

14. **Accepted** - bool (***more straightforward than in 2012***) - but can be nan

15. **MonthlyCost** - only where request accepted

16. **Selected** - only where request accepted

17. **CreditScore** - only where request accepted

18. **OfferedAmount** - only where request accepted

19. **OfferID** - identifies an offer, there are as many unique vals as there are accepted requests
