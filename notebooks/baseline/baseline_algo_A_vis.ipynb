```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_prediction = pd.read_csv("baseline_case2.csv")
```

```python
plt.figure(figsize = (15,8))

ax = sns.violinplot(data = df_prediction, x = "position", y = "baseline_action_pred", jitter=True);
ax.set_xlabel('Position')
ax.set_ylabel('Predicted Action')
ax.set_title('Baseline Predicted Action', size = 20)
```
